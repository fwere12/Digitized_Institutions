```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Reading in our data
data = pd.read_csv("C:\\Users\\Faith\\Downloads\\kenyan_institution_data (2).csv")
data.head(5)
```

```python
# Calculate the percentage of online processes for each organization
data['percentage_online_processes'] = (data['online_process_count'] / data['process_count']) * 100

# Print the updated DataFrame
print(data.head(5))
```

```python
# Replace NaN values with zeros in the percentage_online_processes column
data['percentage_online_processes'].fillna(0, inplace=True)
```

```python
data.head()
```

```python
# Calculate the percentage of online services for each organization
data['percentage_online_services'] = (data['online_service_count'] / data['service_count']) * 100
print(data.head(5))
```

```python
# Replace NaN values with zeros in the percentage_online_processes column
data['percentage_online_services'].fillna(0, inplace=True)
data.head(5)
```

```python
#Add a column called digitization
data['digitization'] = (data['percentage_online_processes'] + data['percentage_online_services']) / 2

# Round the digitization to two decimal places
data['digitization'] = data['digitization'].round(2)

# Display the DataFrame with the new column
print(data)
```

```python
print(data['digitization'].dtype)
```

```python
# Extract the desired columns
columns_to_plot = ['organization_name', 'digitization']
subset_data = data[columns_to_plot]

# Sort the data by 'Digitization' column in descending order
sorted_data = subset_data.sort_values(by='digitization', ascending=False)

# Take the top 10 rows
top_10_data = sorted_data.head(10)

# Plot the horizontal bar chart
plt.barh(top_10_data['organization_name'], top_10_data['digitization'], color='green')
plt.xlabel('Digitization')
plt.ylabel('Organization')
plt.title('Top 10 Most Digitized Institutions', color='red')

# Reverse the order of the y-axis
plt.gca().invert_yaxis()

plt.show()
```
